```python
import cv2
```

```python
img=cv2.imread(r'D:\face_detection\pexels-andrea-piacquadio-712513.jpg')
img
```

```python
cv2.imshow('IMAGE',img)
cv2.waitKey()
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt
plt.imshow(img)
```

```python
conv_img=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(conv_img)
```

```python
face_cascade=cv2.CascadeClassifier(r'D:\face_detection\haarcascade_frontalface_default.xml')
face_cascade
```

```python
img_gray=cv2.cvtColor(conv_img,cv2.COLOR_RGB2GRAY)
img_gray
```

```python
face=face_cascade.detectMultiScale(img_gray,minNeighbors=10)
face
```

```python
len(face)
```

```python
for(x,y,w,h)in face:
    cv2.rectangle(conv_img,(x,y),(x+w,y+h),(0,0,255),6)
```
